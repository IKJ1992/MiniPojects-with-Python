```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('mediaAdvertising.csv')
```

```python
feature = data.iloc[:,:-1]
target = data.iloc[:,-1]
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.25, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(feature_train, target_train)
```

```python
lr.score(feature_train, target_train)
```

```python
lr.coef_
```

```python
feature.columns
```

```python
pd.Series(lr.coef_, index = feature.columns)
```

```python
pred = lr.predict(feature_test)
```
