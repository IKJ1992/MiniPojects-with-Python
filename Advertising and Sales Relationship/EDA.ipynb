```python
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('mediaAdvertising.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
for col in data.columns:
    fig, axes = plt.subplots(1,2, figsize= (20,5))
    sbn.distplot(data[col], ax=axes[0])
    sbn.boxplot(data[col], ax=axes[1])
    plt.show()
    
```

```python
for col in data.columns:
    plt.figure(figsize=(2,1))
    sbn.jointplot(data[col],data['Sales'], kind='reg')
    plt.show()
```

```python
sbn.pairplot(data)
```

```python
data_corr = data.corr()
data_corr
```

```python
sbn.heatmap(data_corr, annot=True, cmap="YlGnBu")
```
