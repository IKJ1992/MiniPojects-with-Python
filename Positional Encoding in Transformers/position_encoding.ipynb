```python
def getPositionEncoding(seq_len, output_dimension, n=10000):
    P = np.zeros((seq_len, output_dimension))
    for k in range(seq_len):
        for i in np.arange(int(output_dimension/2)):
            denominator = np.power(n, 2*i/output_dimension)
            P[k, 2*i] = np.sin(k/denominator)
            P[k, 2*i+1] = np.cos(k/denominator)
    return P

getPositionEncoding(3, 5)
```

```python
PE = getPositionEncoding(50, 500)
plt.matshow(PE)
```
